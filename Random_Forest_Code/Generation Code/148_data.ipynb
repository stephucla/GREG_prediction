```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from collections import Counter

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
movie_dict={}

import os
for dirname, _, filenames in os.walk('/kaggle/input/movie-scripts-corpus/screenplay_data/data/BERT_annotations'):
    for filename in filenames:
        char_dict={}
        cur_speaker=""
        cur_file = open(os.path.join(dirname, filename))
        #to start after 20 lines
        lines=0
        for line in cur_file:
            lines+=1
            if lines>20 and line.startswith("speaker_heading"):
                temp=line[16:].split()
                if temp[0] not in ["(", "(CONT","CONT!","CONTI","CONTINUED", "C"]:
                    cur_speaker=temp[0]
                if cur_speaker not in char_dict:
                    char_dict[cur_speaker]=0
            elif lines>20 and cur_speaker!="" and line.startswith("dialog"):
                temp = line[7:].split()
                char_dict[cur_speaker]+=len(temp)
        sort_char =Counter(char_dict)
        top = sort_char.most_common(3)
        cur_percent=0
        for i in top:
            cur_percent+=i[1]
        cur_file.close()
        total = sum(char_dict.values())
        percent=0
        if (total!=0):
            percent=round((cur_percent/sum(char_dict.values())),6)
        movie_dict[filename]=percent

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
